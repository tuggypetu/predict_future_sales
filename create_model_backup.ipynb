```python
# combine x and y for normalizing

def combine_x_y(x, y):
    ww = []
    for i in tqdm(range(len(x))):
        ggi = [y[i], 0, x[i][2][0], x[i][3][0], x[i][4][0]]
        ww.append(ggi)
    ww = np.array(ww).reshape(214200, 5, 1)
    bg = np.concatenate([x, ww], axis=2)
    return bg
```

```python
combined = combine_x_y(X, Y)
print(combined.shape)
combined
```

```python
c_min = combined.min(axis=(1, 2), keepdims=True)
c_max = combined.max(axis=(1, 2), keepdims=True)

c_n = (combined - c_min)/(c_max - c_min)

print(c_n.shape)
c_n
```
