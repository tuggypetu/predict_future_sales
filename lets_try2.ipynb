```python
def shop_item_tuple(dfn, shop=2, item=3):
    se = []
    num = 0
    fg = None
    for i in tqdm(range(num, dfn.shape[0])):
        ww = f"{int(dfn[i][shop])},{int(dfn[i][item])}"
        se.append(ww)
    se = pd.Series(se, dtype = 'object')
    return se
```

```python
def id_column(dfn, col=0, side=0):
    dfn = np.array(dfn)
    id_list = []
    for r in dfn:
        r = r[col].split(',')
        id_list.append(r[side])
    se = pd.Series(id_list)
    return se
```

```python
def get_0_last_month(df1, df2, col=2):
    check_ids = df2.iloc[:, col].values.tolist()
    col_names = df2.columns
    df1, df2 = df1.values.tolist(), df2.values.tolist()
    df0 = []
    added_ids = []
    for o in tqdm(range(len(df1))):
        _id = df1[o][col]
        if _id not in added_ids and _id not in check_ids:
            row = [f"{_id},33", 0, _id]
            df0.append(row)
            added_ids.append(_id)
    df2 = df2 + df0
    df2 = pd.DataFrame(df2, columns=col_names)
    return df2
```

```python
def create_all_id_month(df1, id_mh_list, col=0):
    col_names = df1.columns
    check_tup = df1.iloc[:, col].values.tolist()
    df1 = df1.values.tolist()
    for p in tqdm(id_mh_list):
        if p not in check_tup:
            _id = p.split(',')[0]
            row = [p, 0, _id]
            df1.append(row)
    df1 = pd.DataFrame(df1, columns=col_names)
    df1 = df1.sort_values('ID')
    df1 = df1.reset_index()
    df1 = df1.drop('index', axis=1)
    return df1
```

```python
def create_tdf_rows(t_df, not_tdf_id):
    rows = []
    for r in not_tdf_id:
        row = [r, 0.0]
        rows.append(row)
    dfu = pd.DataFrame(rows, columns=t_df.columns)
    t_df = pd.concat([t_df, dfu], ignore_index=True)
    return t_df
```

```python
def create_main_array(df1):
    """Create the main array to be used for training"""
    main_list = []
    # df1 = df1.values.tolist()
    for l in tqdm(range(len(df1))):
        tad = df1.ID_data[l].reset_index(drop=True)
        bky = np.array([tad.item_cnt_month.values.astype(int), tad.item_price.values.astype(float), tad.item_category_id.values
                       .astype(int), tad.shop_id.values.astype(int), tad.item_id.values.astype(int)])
        # bky = np.array([tad.item_cnt_month.values.astype(int), tad.item_price.values.astype(float)])
        main_list.append(bky)
    main_array = np.array(main_list).astype(float)
    return main_array
```
